```python
import fitz  # PyMuPDF
from googletrans import Translator
from ipywidgets import widgets
from IPython.display import display

# Function to extract text from a PDF
def pdf_to_text(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text("text") + "\n"
    return text

# Function to translate text
def translate_text(text, dest_language='en'):
    translator = Translator()
    translated = translator.translate(text, dest=dest_language)
    return translated.text

# Upload PDF widget
upload_pdf = widgets.FileUpload(accept='.pdf', multiple=False)
display(upload_pdf)

# Dropdown for language selection
lang_dropdown = widgets.Dropdown(
    options=[('English', 'en'), ('French', 'fr'), ('Spanish', 'es'), ('German', 'de'), ('Hindi', 'hi')],
    value='en',
    description='Translate to:'
)
display(lang_dropdown)

# Button to process PDF
process_button = widgets.Button(description="Convert & Translate")
output = widgets.Output()

def on_button_click(b):
    with output:
        output.clear_output()
        if upload_pdf.value:
            file_info = list(upload_pdf.value.values())[0]
            pdf_path = file_info['metadata']['name']
            with open(pdf_path, "wb") as f:
                f.write(file_info['content'])
            
            # Extract text
            extracted_text = pdf_to_text(pdf_path)
            print("Extracted Text:")
            print(extracted_text[:1000] + "...\n")  # Show first 1000 characters
            
            # Translate text
            translated_text = translate_text(extracted_text, lang_dropdown.value)
            print("\nTranslated Text:")
            print(translated_text[:1000] + "...\n")  # Show first 1000 characters
        else:
            print("Please upload a PDF file.")

process_button.on_click(on_button_click)
display(process_button, output)
```
